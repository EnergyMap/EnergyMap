```python
import geopandas
sweden = geopandas.read_file("ne_10m_admin_0_countries_swe.shp")
```

```python
sweden.plot()
```

```python
sweden.head()
```

```python
swe = sweden[sweden.ADM0_A3_SE == 'SWE']
swe.plot()
```

```python
swe.crs
```

```python
import osmnx as ox
swe1 = ox.project_gdf(swe)
swe1.plot()
```

```python
minx, miny, maxx, maxy = swe1.total_bounds
print(minx, miny, maxx, maxy)
```

```python
from shapely.geometry import Polygon

squares = []

x = minx
y = miny
while x < maxx:
    y = miny
    while y < maxy:
        lon_point_list = [x, x+1000, x+1000, x, x]
        lat_point_list = [y, y, y+1000, y+1000, y]
        squares.append(Polygon(zip(lon_point_list, lat_point_list)))
        y += 1000
    x += 1000

sqdf = geopandas.GeoDataFrame(crs=swe1.crs, geometry=squares)
```

```python
sqswe = sqdf.loc[sqdf.geometry.centroid.within(swe1.loc[50,'geometry'])]
```

```python
sqswe.plot()
```
